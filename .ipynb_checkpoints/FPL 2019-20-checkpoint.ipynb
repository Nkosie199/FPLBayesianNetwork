```python
import numpy as np
import pandas as pd
from scipy import stats

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
week2 = pd.read_excel('./data/week2.xlsx')
week3 = pd.read_excel('./data/week3.xlsx')
week4 = pd.read_excel('./data/week4.xlsx')
week5 = pd.read_excel('./data/week5.xlsx')
week2.info()
```

```python
def plotpointsvsminutes(week):
    area = 1 * week.value
    x = week.minutes
    y = week.total_points
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    print('correlation: ',r)
    #there seems to be a correlation between minutes and total_points
    week.plot.scatter('minutes','total_points',s=area)
    pp.xlabel('Minutes')
    pp.ylabel('Total points')

def plotpointsvsindex(week):
    area = 1 * week.value
    x = week.ict_index
    y = week.total_points
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    print('correlation: ',r)
    #there seems to be a correlation between ict_index and total_points
    week.plot.scatter('ict_index','total_points',s=area)
    pp.xlabel('ICT Index')
    pp.ylabel('Total points')
    
def plotpointsvsinfluence(week):
    area = 1 * week.value
    x = week.influence
    y = week.total_points
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    print('correlation: ',r)
    #there seems to be a correlation between ict_index and total_points
    week.plot.scatter('influence','total_points',s=area)
    pp.xlabel('Influence')
    pp.ylabel('Total points')

def plotpointsvswashome(week):
    #TODO: this requires a histogram not a scatter plot
    data = week.was_home.map({True: 'true', False: 'false'})
    area = 1 * week.value
    week.plot.scatter(data,'total_points',s=area)
    pp.xlabel('Home or away')
    pp.ylabel('Total points')

def plotpointsvsopponent(week):
    area = 1 * week.value
    x = week.opponent_team
    y = week.total_points
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    print('correlation: ',r)
    #there seems to be NO correlation between opponent and total_points
    week.plot.scatter('opponent_team','total_points',s=area)
    pp.xlabel('Opponent team ranking')
    pp.ylabel('Total points')
```

```python
plotpointsvsminutes(week2)
plotpointsvsminutes(week3)
plotpointsvsminutes(week4)
plotpointsvsminutes(week5)
```

```python
plotpointsvsindex(week2)
plotpointsvsindex(week3)
plotpointsvsindex(week4)
plotpointsvsindex(week5)
```

```python
plotpointsvsinfluence(week2)
plotpointsvsinfluence(week3)
plotpointsvsinfluence(week4)
plotpointsvsinfluence(week5)
```
